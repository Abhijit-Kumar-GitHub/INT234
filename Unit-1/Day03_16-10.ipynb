```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../DataSets/ecommerce_customers_unit1.csv')
df
```

```python
df.head()
df.describe()
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
df.isna().sum()
```

```python
df['age'].fillna(df['age'].mean(), inplace = True)
df['total_spent'].fillna(df['total_spent'].mean(), inplace = True)

df.isna().sum()
```

```python

df_numeric = df.select_dtypes(include = [np.number])
print(df_numeric.describe())

numer = df_numeric.columns.to_list()
print(numer)
```

```python
df_norm = df.copy()
norm = MinMaxScaler()

df_norm[numer] = norm.fit_transform(df_norm[numer])
df_norm.info()

```

```python
df_norm.describe()
```

```python
standard = StandardScaler()

df_stan = df.copy()
df_stan[numer] = standard.fit_transform(df_stan[numer])
df_stan.info()
```

```python
# Set the display format for floats
pd.options.display.float_format = '{:.9f}'.format

df_stan.describe()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
le = LabelEncoder()

df_norm['country_encoded'] = le.fit_transform(df_norm['country'])
df_norm
```
