## Polynomial Regression  
> the dependent variable changes non-linearly(exponentially) with independent variable(s) then it is used.  
- still considered linear in model coefficients

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("../Datasets/Position_Salaries.csv")


X = df.iloc[:,1:-1].values
print(X)
y = df.iloc[:,-1].values
print(y)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y);
```

```python
from sklearn.preprocessing import PolynomialFeatures
# creates new features by adding polynomial combinations of the existing features

poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X,lin_reg.predict(X),color = 'blue')
plt.title('Position Salaries using lin regression')
plt.xlabel('Position')
plt.ylabel('Salary')
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color = 'blue')
plt.title('Position Salaries using polynomial regression')
plt.xlabel('Position')
plt.ylabel('Salary')
```

```python
#salary of the person if the level is 15
print(lin_reg.predict([[15]]))
print(lin_reg2.predict(poly_reg.fit_transform([[15]])))
```
